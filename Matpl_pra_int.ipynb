```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x1=np.linspace(0,10,5)
y1=2*x1+3
plt.plot(x1,y1,color='purple',ls='-',lw=3,alpha=0.7,marker='o',
         markersize=15,markeredgecolor='b')
plt.xlabel('x values',fontsize=15)
plt.ylabel('y=2*x+3',fontsize=15)
plt.title('line plot demo',fontsize=20)
plt.legend('Line1',loc='best')
plt.savefig('myfirstgraph1.png',dpi=300)
plt.show()
```

```python
x=np.linspace(0,10,5)
y= 2*x+3
x3=np.arange(0,10,1)
y3= 2*x3+1
y4= -3*x3+2

plt.figure(figsize=(5,5))
plt.subplot(2,1,1)

plt.plot(x3,y3,color='green')
plt.title('graph1')
plt.ylabel('2*x3+1')

plt.subplot(2,1,2)
plt.plot(x3,y4,color='r')
plt.title('graph2')
plt.ylabel('-3*x3+2')
plt.show()
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig= plt.figure(figsize=(5,2))
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
plt.show()
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.2,0.2])
```

```python
fig=plt.figure()
ax1= fig.add_axes([0,0,1,1])
ax2= fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y,color='r')
ax2.plot(x,y,color='r')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.4,.4])
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.4,.4])

ax1.plot(x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('z')
ax2.set_title('Zoom')
```
