```python
#importing the libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1=np.linspace(-3,3,100)
y1=np.sin(x1)
plt.plot(x1,y1)
plt.show()
```

```python
#graphing sigmoid function
x2=np.linspace(-10,10,100)
y2=1/(1+np.exp(-x2))
plt.plot(x2,y2,color='r')
plt.show()
```

```python
x=np.linspace(0,10,50)
y=2*x+3
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(0,10,50)
y=2*x+3
plt.plot(x,y,linestyle='--',color='m',linewidth=2)
plt.xlabel('x values',fontsize=15)
plt.ylabel('y=2*x+3',fontsize=15)
plt.title('line plot demo',fontsize=20)
plt.show()
```

```python
x=np.linspace(0,10,50)
y=2*x+3
plt.plot(x,y,linestyle='-.',color='c',linewidth=3,alpha=0.7)
plt.xlabel('x values',fontsize=15)
plt.ylabel('y=2*x+3',fontsize=15)
plt.title('line plot demo',fontsize=20)
plt.legend('Line1',loc='best')
plt.savefig('myfirstgraph.png')
plt.show()
```

```python
x=np.linspace(0,10,5)
y=2*x+3
plt.plot(x,y,linestyle='-',color='r',linewidth=3,alpha=0.7,marker='o',markersize=15,markeredgecolor='b')
plt.xlabel('x values',fontsize=15)
plt.ylabel('y=2*x+3',fontsize=15)
plt.title('line plot demo',fontsize=20)
plt.legend('Line1',loc='best')
plt.savefig('myfirstgraph1.png',dpi=300)
plt.show()
```

```python
x=np.linspace(0,10,5)
y=2*x+3
x3=np.arange(0,10,1)
y3=2*x3+1
y4=-3*x3+2

plt.figure(figsize=(5,10))
plt.subplot(2,1,1)
#plotting the points
plt.plot(x3,y3)
plt.title('graph1')
plt.ylabel('2*x+1')

plt.subplot(2,1,2)
plt.plot(x3,y4)
plt.title('graph2')
plt.ylabel('-3*x+2')
plt.show()
```
