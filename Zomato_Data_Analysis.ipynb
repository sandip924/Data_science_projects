```python
# check data types
final.dtypes
```

```python
final.Country.value_counts()
```

```python
country_names = final.Country.value_counts().index
country_names
```

```python
country_values = final.Country.value_counts().values
country_values
```

```python
# pie chart- top 3 countries use zomato
plt.pie(country_values[:3], labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

```python
# observation : Zomato maximum records or transactions are from India after that USA and UK
```

```python
final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings=final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

```python
# Observation -1.when rating is 4.5 to 4.9 is excellent  2. rating 4.0 to 4.4 very good from data
```

```python
ratings.head()
```

```python
# matplotlib.rcParams['figure.figsize']=(12,6)
plt.figure(figsize=(12,5))
sns.barplot(x='Aggregate rating',y='Rating count', data=ratings)
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='Aggregate rating',y='Rating count', hue='Rating color',data=ratings)
```

```python
## Observations :- 1. Not rated count is very high 
# 2. maximum number of ratings are between 2.5 to 3.4
```

```python
# countplot 
plt.figure(figsize=(12,5))
sns.countplot(x='Rating color', data=ratings,palette=['blue','red','orange','yellow','green','red'])
```
